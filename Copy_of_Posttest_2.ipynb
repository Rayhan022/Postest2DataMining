Dan buatlah daftar support dan confidence untuk 2 item set dengan menggunakan dataset tersebut (hint : buatlah tabel tabulasi terlebih dahulu untuk memudahkan)

```python
import pandas as pd
```

```python
df1 = pd.read_excel("Dataset_baru.xlsx",sheet_name="Sheet1")
```

```python
df1
```

```python
df1.sum()
```

```python
df1.columns
```

```python
#library
from itertools import permutations
```

```python
#menyimpan kombinasi
rules = list(permutations(df1.columns,2))
```

```python
#menampilkan
rules
```

```python
#simpan dalam bentuk list
rule_baru = []
#perulangan dataset
for rule in rules:
  #mengetahui kombinasi ada atau tidak
  if len(df1[(df1[rule[0]]==1) & (df1[rule[1]]==1)])>0:
    #menyimpan dalam bentuk tuple
    rule_baru.append(rule)
rule_baru
```

```python
#menyimpan dalam dictionary
hitung = {} 
for rule in rule_baru:
  banyak_data= len(df1[
      (df1[rule[0]]==1) & (df1[rule[1]]==1)
      ])
  hitung[str(rule)] = banyak_data

hitung
```

```python
#perulangan setiap kombinasi
for item,ab in hitung.items():
  #penghitungan pada kombinasi
  support = ab/len(df1)*100
  print(f"{item} support = {support:3f}%")
```

```python
for item,ab in hitung.items():
  #"A->B"
  i = eval(item) #Ubah ke list dari string
  print(f"{i[0]}->{i[1]}  Confidence = {ab/df1[i[0]].sum()*100}%")
```
